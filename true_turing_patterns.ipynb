```python
import numpy as np
import matplotlib.pyplot as plt

from turing import generate_pattern
from drawing import maps_img
```

```python
!mkdir turing_224_true
```

```python
img_size = (224, 224)
N = 13

for l1 in range(1, N, 2):
    for l2 in range(1, N, 2):
        pattern = generate_pattern(img_size, kernel_N=N, kernel_L1=l1, kernel_L2=l2)
        np.save(f'turing_224_true/turing_{l1}_{l2}.npy', pattern)
```

```python
fig, ax = plt.subplots(N // 2, N // 2, figsize=(60, 60))

for l1 in range(1, N, 2):
    for l2 in range(1, N, 2):
        pattern =  np.load(f'turing_224_true/turing_{l1}_{l2}.npy')
        
        i = l1 // 2
        j = l2 // 2
        ax[i][j].imshow(maps_img(pattern))
        ax[i][j].axis('off')
        ax[i][j].set_title(f'L1={l1}, L2={l2}')
```
